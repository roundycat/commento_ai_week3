```python
from ultralytics import YOLO

# 사전 학습된 YOLO11n-seg 모델 로드
model = YOLO("yolo11n-seg.pt")

# 데이터 증강을 적용하여 학습 (20 Epoch)
results = model.train(
    data="carparts-seg.yaml",  # 데이터셋 설정
    epochs=3,                 # 학습 반복 횟수
    imgsz=640,                  # 입력 이미지 크기
    augment=True                # 데이터 증강 활성화
)
```

```python
import glob
import cv2
from ultralytics import YOLO

# runs/segment/ 아래에서 가장 최근의 best.pt 찾기
latest_model = sorted(glob.glob("runs/segment/*/weights/best.pt"))[-1]
print("Loaded model:", latest_model)

model = YOLO(latest_model)

# 테스트 이미지 불러오기
image_path1 = "datasets\\carparts-seg\\images\\test\\car4_jpg.rf.8978131a7b03be689c244641e42e1307.jpg"
image_path2 = "datasets\\carparts-seg\\images\\train\\car5_jpg.rf.1134bd6ff911228406b6312db4a0e4ff"

image1 = cv2.imread(image_path1)

# 객체 탐지 및 분할 실행
results = model(image1)

# 탐지 결과 시각화
for result in results:
    masks = result.masks  # 세그멘테이션 마스크
    boxes = result.boxes  # 바운딩 박스
    names = result.names

    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        label = names[int(box.cls[0])]
        confidence = box.conf[0]
        cv2.rectangle(image1, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image1, f"{label} ({confidence:.2f})", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# 결과 출력
cv2.imshow("Carparts Segmentation", image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
metrics = model.val()
print(metrics)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 1) CSV 경로 (본인 경로로 수정 가능)
csv_path = Path("runs/segment/train2/results.csv")

# 방어 코드: 경로 체크
if not csv_path.exists():
    raise FileNotFoundError(f"CSV가 없어요: {csv_path}")

# 2) CSV 로드 + 컬럼명 공백 제거
df = pd.read_csv(csv_path)
df.columns = [c.strip() for c in df.columns]

# 디버깅용: 어떤 컬럼들이 있는지 한번 보려면 주석 해제
# print(df.columns.tolist())

# 3) 원하는 컬럼을 "키워드 포함"으로 탐색하는 헬퍼
def find_col(*keywords):
    for c in df.columns:
        if all(k in c for k in keywords):
            return c
    return None

# precision / recall (Box 기준) 후보 찾기
prec_col   = find_col("metrics/precision")   # 예: 'metrics/precision(B)'
recall_col = find_col("metrics/recall")      # 예: 'metrics/recall(B)'

# mAP50(Box), mAP50-95(Box) 후보
map50_box_col    = (find_col("metrics/mAP50", "(B)") or find_col("metrics/mAP50"))
map5095_box_col  = (find_col("metrics/mAP50-95", "(B)") or find_col("metrics/mAP50-95"))

# mAP50(Mask), mAP50-95(Mask) 후보 (세그멘테이션용)
map50_mask_col   = find_col("metrics/mAP50", "(M)")
map5095_mask_col = find_col("metrics/mAP50-95", "(M)")

# 4) 있는 것만 그리기
plotted = False

if prec_col and recall_col:
    plt.figure()
    plt.plot(df[prec_col], label="Precision (Box)")
    plt.plot(df[recall_col], label="Recall (Box)")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.title("Precision / Recall over Epochs")
    plt.legend()
    plt.show()
    plotted = True

if map50_box_col or map5095_box_col:
    plt.figure()
    if map50_box_col:
        plt.plot(df[map50_box_col], label="mAP50 (Box)")
    if map5095_box_col:
        plt.plot(df[map5095_box_col], label="mAP50-95 (Box)")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.title("Box mAP over Epochs")
    plt.legend()
    plt.show()
    plotted = True

if map50_mask_col or map5095_mask_col:
    plt.figure()
    if map50_mask_col:
        plt.plot(df[map50_mask_col], label="mAP50 (Mask)")
    if map5095_mask_col:
        plt.plot(df[map5095_mask_col], label="mAP50-95 (Mask)")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.title("Mask mAP over Epochs")
    plt.legend()
    plt.show()
    plotted = True

if not plotted:
    print("그릴 수 있는 컬럼을 못 찾았어요. 아래 컬럼 리스트를 확인하고 키워드를 조정하세요:")
    print(df.columns.tolist())
```
